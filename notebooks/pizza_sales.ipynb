```python
df_cat = df.groupby("pizza_category")["unit_price"].agg(["mean", "min", "max", "median", "count"]).round(2)
df_cat
```

```python
df_cat = df.groupby("pizza_name")["unit_price"].agg(["mean", "min", "max", "median", "count"]).round(2).sort_values(by="count", ascending=False)
df_cat
```

```python
print(df.columns)
df_orders = pd.DataFrame(df[["order_id", "order_date" ,"order_time"]].groupby("order_id").value_counts())
df_orders.drop(columns=["count"], inplace=True)
df_orders.to_csv('orders.csv')
df_orders
```
