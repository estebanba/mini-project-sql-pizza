```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Sizes Chart

df = pd.read_csv('../data/charts/size_quantity_revenue.csv')

plt.figure(figsize=(12, 6))

plt.bar(df['size_name'], df['quantity_percentage'], color='lightblue', label='Quantity %')
plt.bar(df['size_name'], df['revenue_percentage'], alpha=0.5, color='coral', label='Revenue %')
plt.plot(df['size_name'], df['revenue_quantity_ratio'] * 10, color='black', marker='o', 
         linewidth=2, label='Revenue/Quantity Ratio * 10')

plt.title('Size Performance Metrics', pad=20)
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Categories Chart

df = pd.read_csv('../data/charts/category_quantity_revenue.csv')

plt.figure(figsize=(12, 6))

plt.bar(df['pizza_category'], df['quantity_percentage'], color='lightblue', label='Quantity %')
plt.bar(df['pizza_category'], df['revenue_percentage'], alpha=0.5, color='coral', label='Revenue %')
plt.plot(df['pizza_category'], df['ratio_revenue_quantity'] * 10, color='black', marker='o', 
         linewidth=2, label='Revenue/Quantity Ratio * 10')

plt.title('Category Performance Metrics', pad=20)
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Top 5 Pizzas Chart
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/charts/top5_pizzas_revenue.csv')

plt.figure(figsize=(12, 6))

plt.bar(df['name'], df['quantity_percentage'], color='lightblue', label='Quantity %')
plt.bar(df['name'], df['revenue_percentage'], alpha=0.5, color='coral', label='Revenue %')
plt.plot(df['name'], df['revenue_quantity_ratio'] * 10, color='black', marker='o', 
         linewidth=2, label='Revenue/Quantity Ratio * 10')

plt.title('Top 5 Pizzas Performance Metrics', pad=20)
plt.xticks(rotation=45, ha='right')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

```
